```python
import pandas as pd
import matplotlib.pyplot as plt

# Učitaj CSV fajlove
df_linear = pd.read_csv('prosjek_linear.csv')
df_lstm = pd.read_csv('prosjek_lstm.csv')

# Pretpostavljam da oba imaju kolone: 'Godina' i 'Prosjek'

plt.figure(figsize=(12,6))

plt.plot(df_linear['Godina'], df_linear['Prosjek'], label='Linearna regresija', marker='o')
plt.plot(df_lstm['Godina'], df_lstm['Prosjek'], label='LSTM model', marker='o')

plt.title('Prosječne godišnje temperature (2026-2045)')
plt.xlabel('Godina')
plt.ylabel('Temperatura (°C)')
plt.xticks(df_linear['Godina'], rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
